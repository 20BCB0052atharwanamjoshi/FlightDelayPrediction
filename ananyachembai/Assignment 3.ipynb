```python
from scipy import stats

z_scores = stats.zscore(data['price'])

threshold = 3
outliers = np.abs(z_scores) > threshold

data = data[~outliers]
```

```python
plt.scatter(x=data.index, y=data['price'])

plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Scatter Plot of Price')
```

```python
#Task 7- Categorical columns encoding
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
data['mainroad'] = le.fit_transform(data['mainroad'])
data['guestroom'] = le.fit_transform(data['guestroom'])
data['basement'] = le.fit_transform(data['basement'])
data['hotwaterheating'] = le.fit_transform(data['hotwaterheating'])
data['airconditioning'] = le.fit_transform(data['airconditioning'])
data['furnishingstatus'] = le.fit_transform(data['furnishingstatus'])
```

```python
data.info()
```

```python
data.head()
```

```python
#Task 8- Dependant and independant variables

y = data['mainroad']

x = data.drop('mainroad', axis=1)
```

```python
x
```

```python
y
```

```python
#Task 9- Scale the independent variables
```

```python
from sklearn.preprocessing import StandardScaler

name = x.columns
scale = StandardScaler()
x = scale.fit_transform(x)
x
```

```python
x = pd.DataFrame(x, columns=name)
x
```

```python
#Task 10- Splitting the data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=0)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#Task 11- Building the model

from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
```

```python
#Task 12- Training the model

lor.fit(x_train, y_train)
```

```python
#Task 13- Testing the model

y_pred = lor.predict(x_test)
```

```python
y_pred
```

```python
#Task 14- Performance metrics
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
roc_auc_score(y_test, y_pred)
```

```python
fpr, tpr, threshold = roc_curve(y_test, y_pred)

plt.plot(fpr, tpr, marker='.')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```
